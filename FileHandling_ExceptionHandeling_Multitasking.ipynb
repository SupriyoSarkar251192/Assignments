6. Explain the concept of buffering in file handeling and how it helps in improving read and write operations?

Ans: Buffering is a technique that Python uses to improve the efficiency of file handling operations. When we open a file in Python, the file object is associated with a buffer, which is a region of memory that stores data temporarily. When we read or write data to the file, Python transfers data between the buffer and the file object.

There are three different buffering modes in Python:

Unbuffered mode:
In unbuffered mode, Python transfers data between the buffer and the file object one byte at a time. This is the slowest buffering mode, but it is also the most reliable.

Line buffered mode:
In line buffered mode, Python transfers data between the buffer and the file object one line at a time. This is faster than unbuffered mode, but it is less reliable because Python may not flush the buffer to the file object until the newline character is encountered.

Buffered mode:
In buffered mode, Python transfers data between the buffer and the file object in chunks of data. This is the fastest buffering mode, but it is also the least reliable because Python may not flush the buffer to the file object until the buffer is full or the file object is closed.

The default buffering mode for file objects in Python is buffered mode. You can change the buffering mode for a file object by passing the buffering argument to the open() function. 
The buffering argument can be one of the following values:
0: Unbuffered mode
1: Line buffered mode
Any other value: Buffered mode

4. Close the BufferedReader object.

```python
# 8. Write a python function to read a text file using buffered reading and return its contents.
from io import BufferedReader
def buff_read(x):
    with open(f"{x}", "rb") as bf_read:
        with BufferedReader(bf_read) as reader:
            contents = reader.read()
    return contents

file_name4 = input("insert file name:")
print(buff_read(file_name4))
```

```python
# 25. Write a try except finally block to handle a value error.
list_val = [4, 3, 7, 'l', 9, 5]
i = 0
try:
    while i < len(list_val):
        print(float(list_val[i]))
        i += 1
except ValueError as e:
    print(e)
finally:
    while i < len(list_val) - 1:
        i += 1
        print(float(list_val[i]))
        
```

For example, if we are creating a banking application, we may want to define an exception that is raised when a user tries to withdraw more money than they have in their account. This exception can be defined as a custom exception, such as InsufficientFundsError, which makes it easier to handle this specific error condition.

```python
# 28. Create a custom exception class with a message.
class ImproperAge(Exception):
    def __init__(self, age, msg1 = "invalid age", msg2 = "not eligible for vote"):
        self.age = age
        self.msg1 = msg1
        self.msg2 = msg2
        if self.age < 0 or self.age > 100:
            super().__init__(self.msg1)
        if self.age >= 0 and self.age < 18:
            super().__init__(self.msg2)
candidate_age = int(input())
try:
    if candidate_age >= 18 and candidate_age <= 100:
        print("eligible to vote.")
    else:
        raise ImproperAge(candidate_age)
except ImproperAge as e:
    print(e)
```

```python
# 29. Write a code to raise a custom exception error.
class IncompleteTransaction(Exception):
    def __init__(self):
        super().__init__("Insufficient balance.")
class banking:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, amt1):
        self.balance = self.balance + amt1
        print(f"Transaction successful. New balance: Rs. {self.balance}")
    def withdraw(self, amt2):
        try:
            if amt2 <= self.balance:
                self.balance = self.balance - amt2
                print(f"Transaction successful. New balance: Rs. {self.balance}")
            else:
                raise IncompleteTransaction
        except IncompleteTransaction as e:
            print(e)
Supriyo = banking(35000)
Supriyo.deposit(7000)
Supriyo.withdraw(50000)
Supriyo.withdraw(40000)
Supriyo.withdraw(5000)
```

```python
# 30. Write a code to raise a custom exception when a value is negative.
class NegativeNumber(Exception):
    def __init__(self):
        super().__init__("Negative number not accepted. Enter a positive integer.")
def factorial(n):
    try:
        if n == 0 or n == 1:
            return 1
        elif n > 1:
            fact = n * factorial(n - 1)
            return fact
        else:
            raise NegativeNumber
    except NegativeNumber as e:
        return e
n1 = int(input())
print(factorial(n1))
```

33. What is multithreading?

Ans: Multithreading in Python is a way to achieve multitasking by executing multiple threads concurrently. A thread is a lightweight process that can run independently of other threads. Threads share the same memory space as the main process, but they each have their own stack. This allows threads to communicate with each other and share data easily.

```python
# 34. Create a thread in python.
import threading
def table(x, y):
    print(f"{x} * {y} = {x * y}")
n = int(input())
table_threads = [threading.Thread(target=table, args=(n, i)) for i in range(1, 11)]
for t in table_threads:
    t.start()
```

35. What is Global Interpreter Lock in Python?

Ans: The Global Interpreter Lock (GIL) is a mechanism in the CPython interpreter that allows only one thread to execute Python bytecode at a time. This means that even if you have a multi-core processor, only one core can be used to execute Python code at a time.
The GIL is in place to ensure thread safety and prevent race conditions. Race conditions occur when two or more threads are trying to access the same data at the same time and can lead to data corruption. By only allowing one thread to execute Python bytecode at a time, the GIL prevents race conditions from happening.

```python
# 36. Implement a simple multithreading example in python.
import threading
def palindrom(n):
    str_num = str(n)
    str_num1 = str_num[::-1]
    if str_num == str_num1:
        print(n, end="  ")
r = int(input("input range: "))
pal_threads = [threading.Thread(target=palindrom, args=(i,)) for i in range(r + 1)]
for t in pal_threads:
    t.start()
```

However, multiprocessing can be less efficient for I/O-bound tasks, because the processes must communicate with each other through inter-process communication (IPC) mechanisms, which can be slower than accessing shared memory.

43. Explain inter-process communication in multiprocessing.

Ans: Inter-process communication (IPC) is a mechanism that allows processes to communicate with each other. In multiprocessing, IPC is used to allow multiple processes to share data and coordinate their activities.
There are a number of different IPC mechanisms available, including:
Pipes: Pipes are a simple way to allow two processes to communicate with each other. One process writes to the pipe, and the other process reads from the pipe.
Queues: Queues are a more sophisticated IPC mechanism that allows multiple processes to communicate with each other. Processes can add data to the queue, and other processes can remove data from the queue.
Shared memory: Shared memory allows processes to share a common region of memory. This can be used to allow processes to communicate with each other by reading and writing to the shared memory.
Semaphores: Semaphores are a synchronization mechanism that can be used to allow processes to coordinate their activities.
Sockets: Sockets are a network IPC mechanism that allows processes to communicate with each other over a network.
The choice of IPC mechanism depends on the specific needs of the application. For example, pipes are a good choice for simple communication between two processes, while queues are a good choice for more complex communication between multiple processes.
