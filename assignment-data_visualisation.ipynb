```python
from bokeh.plotting import figure, output_file, show
import pandas as pd
import numpy as np
```

```python
"""1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

"""
data = {
    'x': np.array([i for i in range(11)]),
    'y': np.array([np.sin(i) for i in range(11)])
}
df = pd.DataFrame([data])

x = list(range(0, 11))
y = [np.sin(i) for i in x]
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="y")
p.line(x, y, legend_label="Line", line_width=2)
show(p)
```

```python
"""2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

"""

x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]
p = figure(title="Scatter Plot", x_axis_label="x", y_axis_label="y")
p.circle(x,y0,legend_label="Circles",fill_color="red",fill_alpha=0.5,line_color="blue",size=80,)
p.square(x,y1,legend_label="Squares",fill_color="blue",fill_alpha=0.5,line_color="red",size=60,)
p.triangle(x,y2,legend_label="Triangles",fill_color="green",fill_alpha=0.5,line_color="red",size=40)
show(p)
```

```python
"""3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
"""
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
"""4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

"""

p = figure(width=670, height=400, toolbar_location=None, title="Histogram")

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="skyblue", line_color="white",
         legend_label="1000 random samples")
show(p)
```

```python
"""5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
"""

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y")
p.image(image=[data_heatmap], x=[0], y=[0], dw=1, dh=1, palette="Viridis256")
show(p)
```
